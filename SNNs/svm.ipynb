```python

data = np.load(r"/home/sameerahtalafha/Downloads/5Cdata/data5C.npy") # 5 channels data,(num,100,100,5), channel 0-4 is DEM,RGB,NIR
labels = np.load(r"/home/sameerahtalafha/Downloads/5Cdata/label.npy")
data=data[:,:,:,:1]

data1=[]
k=np.zeros((100,100,1))
k=k.astype('float32')
for i in range (len(data)):
    if np.array_equal(k,data[i]) ==True:
        print(i)
    else:
        data1.append(data[i])
index = [306,417,418,528,529,924,935,1000,1051,1246,1247,1354,1441,1578,1579,1580,1712,1807,1808]
labels = np.delete(labels, index)
     

data1=np.array(data1)
data1=data1.reshape(data1.shape[0],1,100,100)
x_train,x_test,y_train,y_test = train_test_split(data1,labels,test_size=0.2,random_state=1,stratify=labels)
```
